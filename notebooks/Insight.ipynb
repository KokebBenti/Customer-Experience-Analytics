```python
import pandas as pd
df1=pd.read_csv("dashen_clean.csv")
df2=pd.read_csv("boa_clean.csv")
df3=pd.read_csv("cbe_clean.csv")
```

```python
df1.head()
```

```python
import pandas as pd
dfc=pd.concat([df1,df2,df3],axis=0,ignore_index=True)
counts=dfc.groupby('bank')["sentiment(textblob)"].value_counts().unstack(fill_value=0)
import matplotlib.pyplot as plt
counts.plot(kind='bar',stacked=False)
plt.show()
```

```python
from itertools import chain
from wordcloud import WordCloud
positive_reviews = df2[df2['sentiment(textblob)'] == 'positive']['clean review']
negative_reviews = df2[df2['sentiment(textblob)'] == 'negative']['clean review']
negative_text = ' '.join(negative_reviews)
neg_wc = WordCloud(width=800, height=400, background_color='white').generate(negative_text)

plt.figure(figsize=(10, 5))
plt.imshow(neg_wc, interpolation='bilinear')
plt.axis('off')
plt.title("Negative Review Word Cloud(BOA)")
plt.show()
```
